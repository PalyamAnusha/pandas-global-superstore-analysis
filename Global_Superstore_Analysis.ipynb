```python
# Sales Data Analysis with Pandas

This project demonstrates how to create, analyze, and visualize a sales dataset using Python, Pandas, and Seaborn.

## 1. Import Libraries
We'll start by importing the required Python libraries.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## 2. Create a Sample Sales Dataset
We'll create a dummy dataset to simulate sales records.
```

```python
import pandas as pd

data = {
    'Date': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'Product': ['Laptop', 'Tablet', 'Phone', 'Laptop', 'Tablet', 'Phone', 'Laptop', 'Tablet', 'Phone', 'Laptop'],
    'Units_Sold': [5, 3, 8, 2, 7, 1, 6, 4, 9, 3],
    'Price_per_Unit': [1000, 500, 800, 1000, 500, 800, 1000, 500, 800, 1000]
}

df = pd.DataFrame(data)
df['Total_Sales'] = df['Units_Sold'] * df['Price_per_Unit']
df
```

```python
## 3. Data Analysis
Let's check basic statistics and patterns in the sales data.
```

```python
print("Basic Statistics:\n", df.describe())
print("\nSales by Product:\n", df.groupby('Product')['Total_Sales'].sum())
```

```python
## 4. Data Visualization
We'll visualize the total sales trend and sales by product.
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(10,5))
sns.lineplot(data=df, x='Date', y='Total_Sales', marker='o')
plt.title("Total Sales Over Time")
plt.show()

sns.barplot(data=df, x='Product', y='Total_Sales', estimator=sum)
plt.title("Total Sales by Product")
plt.show()
```
